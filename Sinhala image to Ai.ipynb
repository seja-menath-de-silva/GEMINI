```python

!pip install gtts
!pip install googletrans==4.0.0-rc1
!pip install google.colab
!pip install gtts googletrans google.colab


!pip install gtts googletrans==4.0.0-rc1 google.colab

from google.colab import files
from PIL import Image
from io import BytesIO
import google.generativeai as genai
from IPython.display import display, Markdown
from gtts import gTTS
from IPython.display import Audio
from googletrans import Translator

# Configure the API key for generative AI
genai.configure(api_key="AIzaSyAgYD9komBIepaqDvKT3FJSVbynsc9WVkg")

# Set up the model and configuration
generation_config = {
  "temperature": 0.4,
  "top_p": 1,
  "top_k": 32,
  "max_output_tokens": 4096,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  }
]

model = genai.GenerativeModel(model_name="gemini-pro-vision",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

# Upload image file using Colab's file upload
uploaded = files.upload()

# Process the uploaded image
for img_name, img_data in uploaded.items():
    img = Image.open(BytesIO(img_data))
    img_byte_array = BytesIO()
    img.save(img_byte_array, format='JPEG')
    image_parts = [
        {
            "mime_type": "image/jpeg",
            "data": img_byte_array.getvalue()
        },
    ]

    user_input = input("Enter your prompt: ")  # Asking for user input

    prompt_parts = [
        image_parts[0],
        f"\n {user_input} ",
    ]

    response = model.generate_content(prompt_parts)

    display(Markdown(f"###✨️{response.text}[ © Developer : seja menath de silva ] "))  # Displaying the result text in bold
    break  # Exiting the loop after the first execution

for img_name, img_data in uploaded.items():
    img = Image.open(BytesIO(img_data))
    display(img)  # Displaying the image
    break

# Get the generated response text
response_text = response.text

# Create a translator object
translator = Translator()

# Translate the text to French
translated_text = translator.translate(response_text, dest='si').text

# Create a gTTS object for Sinhala language
tts = gTTS(text=translated_text, lang='si')  # Assuming the text is translated to Sinhala

# Save the generated speech as an MP3 file
tts.save('output.mp3')

# Display and play the audio in Colab
Audio('output.mp3', autoplay=True)
```
